# Implementación de funciones

Implemetación impulso unitario:

```python
def impseq(n0,n1,n2):
  # Genera x(n) = delta(n-n0); n1<=n<=n2

  n = np.arange(n1,n2+1) #se crea el vector de muestras
  x= (n-n0)== 0
  return [x,n]

x, n1=impseq(2,1,5)

plt.stem(n1,x)
```

1. Implementación escalón unitario:

```python
def escuni(n0,n1,n2):

  n = np.arange(n1,n2+1) #Se crea el vector de muestras
  x = (n-n0) >= 0

  return [x,n]


x2, n2 = escuni(5,1,10)
plt.stem(n2,x2)
```

2. Implementación función rampa:

```python
def rampa(n0,n1,n2,m):

  n = np.arange(n1,n2+1) #se crea el vector de muestras
  x= ((n-n0) >= 0)*(n-n0)*m
  return [x,n]

x3, n3 = rampa(2,1,10,1)
plt.stem(n3,x3)
```

# Generación de secuencias

3. a) Secuencia x1(n)

  x1(n) = 3 * δ(n+1) + 5 * δ(n+3) + 3 * δ(n+2) + 3 * δ(n) + δ(n)

```python
# El vector de las muestras debe ser... iniciar en -3 y terminar en 1

x1, n=impseq(-1,-3,0)
x2, n=impseq(-3,-3,0)
x3, n=impseq(-2,-3,0)
x4, n=impseq(0,-3,0)
x5, n=impseq(0,-3,0)
s= 3*x1+5*x2+3*x3+3*x4+x5

plt.stem(n,s)
plt.xlabel("n")
plt.ylabel("x[n]")
plt.title("Secuencia x1(n)")
```

3. b) Secuencia x2(n) con -6 < n < 6

  x2(n) = -2 * u(n+4) + 3 * u(n+2)

```python
x1, n=escuni(-4,-6,6)
x2, n=escuni(-2,-6,6)

r= (-2*x1)+(3*x2)
plt.stem(n,r)
plt.xlabel("n")
plt.ylabel("x[n]")
plt.title("Secuencia x2(n)")
```

3. c) Secuencia x3(n) con 0 <= n <= 100

```python
n = np.arange(0,101)
x3 = np.exp(0.01)*np.sin(2*np.pi*n)
plt.stem(n,x3)
plt.xlabel("n")
plt.ylabel("x[n]")
plt.title("Secuencia x3(n)")
```

3. d) Secuencia x4(n) con -10 <= n <= 10

  x4(n) = 2 * r(n+3) - r(n-2) - 5 * u(n-3)


```python
x1, n = rampa(-3,-10,10,1)
x2, n = rampa(2,-10,10,1)
x3, n = escuni(3,-10,10)

x4 = 2*x1-x2-5*x3
plt.stem(n,x4)
plt.xlabel("n")
plt.ylabel("x[n]")
plt.title("Secuencia x4(n)")
```
