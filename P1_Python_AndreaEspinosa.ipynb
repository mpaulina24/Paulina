```python
#Se crean la señal coseno y seno con un n en un intervalo entre 0 y 80.
n = np.linspace(0, 80, 650)
y= np.sin(np.pi*0.18*n)
y2= np.cos(2*np.pi*0.03*n)

#Se crea dos nuevas señales a partir de y y y2, siendo la suma y multiplicación de estas respectivamente
s= y + y2
t = y*y2
```

```python
#Se grafica la señal y  y y2 en la misma gráfica
plt.figure()
plt.plot(n,y, label="y =  sin(pi*0.18*n)", color = 'k')
plt.plot(n, y2, label="y2 =cos(2*pi*0.03*n)", color ='m')
plt.title("Señales")
plt.xlabel("n")
plt.ylabel("Amplitud")
plt.legend()
plt.grid(True)
plt.show()
```

```python
#Se gráfica las señales s y t en la misma gráfica
plt.figure()
plt.plot(n,s, label="s = y+y2", color = 'b')
plt.plot(n,t, label="t = y*y2", color ='g')
plt.title("Señales")
plt.xlabel("n")
plt.ylabel("Amplitud")
plt.legend()
plt.grid(True)
plt.show()
```

```python
def funcion_notas(diccionario):

    nota_max= max(diccionario.values())
    nota_min= min(diccionario.values())
    nota_media = statistics.mean(diccionario.values())
    dev_tipica = statistics.stdev(diccionario.values())


    notas_serie = pd.Series({'nota máxima': nota_max, 'nota mínima': nota_min, 'nota media': nota_media, 'desviación tipica': dev_tipica})
    return print(notas_serie)

notas= {'A1':3.3, 'A2':4.5, 'A3':5.0, 'A4': 2.2, 'A5':3.0, 'A6': 3.9, 'A7':4.3}
funcion_notas(notas)
```
