```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import pandas as pd
```

```python
a = np.array([67.1, 1,-0.3,5.2,-6]) #vector a
print("Vector a: ", a)

b = np.array([1,3,2.2,5.1,1]) #vector b
print("\nVector b: ", b)
```

```python
"""Para hacer el producto punto se tiene en cuenta que ambos vectores creados posean la misma dimensión, el orden de los
vectores no importa ya que este producto conmuta. Se puede hacer de dos formas:"""

# 1. con la función directa de python
print("a.b (producto punto): ", a.dot(b))

# 2. Con la función de Numpy, en la que se pasan los dos vectores como argumentos
print("\na.b (producto punto): ", np.dot(a,b))

print("\na*b = ", np.multiply(a,b)) #multiplicar uno a uno los vectores
```

```python
A = np.array([[2, -1, -3], [4, -1.5, -2.5],[7.3,-0.9,-0.2]]) #Creación de la matriz A
print("Matriz A: \n",A)
```

```python
c=np.transpose(A) #Sacar la transpuesta de la matriz
print("Transpuesta de A: \n",c) #Imprimir transpuesta de A

print("\nPosición A(0,2): ", A[(0,2)]) #fila 1, columna 3 de la matriz A

print("\nSegunda fila de A: ", A[1,:]) #obtener la segunda fila de la matriz A

print("\nDimensiones de A: ", np.shape(A)) #Conocer dimensiones de la matriz A
```

```python
# Ejemplos de funciones

# numpy.ones(): devuelve un array del tamaño y tipo indicados inicializando sus valores con unos
print("Ejemplo función ones: \n", np.ones((2, 5), dtype = 'int'))

# numpy. round() se utiliza para redondear un valor al entero más cercano con una cantidad de decimales indicada si se requiere
print("\nEjemplo redondeo hacia el número más cercano:", round(3.3863,2))

# numpy. ceil() redondea un número hacia arriba al entero más cercano o igual.
print("\nEjemplo redondeo hacia arriba de 3.3: ", np.ceil(3.3))

# numpy. floor() redondea un número hacia abajo al entero más cercano o igual.
print("\nEjemplo redondeo hacia abajo de 3.3: ", np.floor(3.3))
```

```python
#Creación de señales y graficación

n = np.linspace(0, 80, 800) #Construcción del vector que va de 0 a 80 para graficar las funciones yn, y2n

yn = np.sin(np.pi*0.18*n) #Función 1
y2n = np.cos(2*np.pi*0.03*n) #Función 2

s = np.add(yn,y2n) #Suma de funciones

t = np.multiply(yn,y2n) #Multiplicación de funciones

#Graficación
plt.figure(1)
plt.plot(n, yn, '-g', label = "yn")
plt.plot(n, y2n, ':b', label = "y2n")
plt.legend()
plt.title("Gráfico 1")
plt.xlabel("n")
plt.ylabel("Amplitud")
plt.grid(True)

plt.figure(2)
plt.plot(n, s, '-g', label = "s = yn + y2n")
plt.plot(n, t, ':b', label = "t = yn * y2n")
plt.legend()
plt.title("Gráfico 2")
plt.xlabel("n")
plt.ylabel("Amplitud")
plt.grid(True)
```

```python
#Reto función

#Creación de la función
def funcion(dict1):

  datos = pd.Series(dict1)
  c= np.array(datos["Carlos":"Juan"]) #Creación de una lista con las notas del primer al último nombre del diccionario

  #Cálculos de promedio, valor mínimo, máximo y desviación típica
  prom = np.mean(c)
  min = np.min(c)
  max = np.max(c)
  ds = c.std()

  serie = pd.Series({"Promedio":prom,"Mínimo":min,"Máximo":max,"Desv típica":ds}) #Creación de la serie deseada

  return serie

dict1 = {"Carlos":5,"Maria":3,"Omar":1,"Ana":2,"Juan":1} #Diccionario con nombres y notas

funcion(dict1) #Llamar la función
```
